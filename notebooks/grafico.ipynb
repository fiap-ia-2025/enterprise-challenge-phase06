```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data/medicao_vibracao.csv')
```

```python
# Mapeamento de cores
cor_mapa = {
    'NORMAL': 'green',
    'ALERTA_Pre_falha': 'orange',
    'FALHA_CRITICA': 'red'
}
cores = df['Status'].map(cor_mapa)

# Grfico de barras
plt.figure(figsize=(16, 6))
plt.bar(df['ID'], df['Aceleracao_Total'], color=cores, width=0.5)

plt.xlabel('Medies')
plt.ylabel('Acelerao Total')
plt.title('Acelerao por status ao longo das medies')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Legenda
import matplotlib.patches as mpatches
legenda = [mpatches.Patch(color=v, label=k) for k, v in cor_mapa.items()]
plt.legend(handles=legenda)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Carrega o arquivo de temperatura para o DataFrame 'df'
# Usamos o caminho que j corrigimos
df = pd.read_csv('../enterprise-challenge-phase04-main/data/medicao_temperatura.csv')

# Limpa os espaos de TODOS os nomes de colunas de uma vez
df.columns = df.columns.str.strip()

# Mapeamento de cores, agora usando o nome limpo 'Status'
cor_mapa = {
    'NORMAL': 'green',
    'ALERTA_Pre_falha': 'orange',
    'FALHA_CRITICA': 'red'
}
cores = df['Status'].map(cor_mapa) # Usando a coluna 'Status' limpa

# Grfico de barras
plt.figure(figsize=(16, 6))

# Usar o DataFrame 'df' e as colunas com nomes limpos ('ID', 'Temperatura')
plt.bar(df['ID'], df['Temperatura'], color=cores, width=0.5)

# Configurao do Grfico (seu cdigo aqui j estava timo)
plt.xlabel('Medies')
plt.ylabel('Temperatura (C)')
plt.title('Temperatura por status ao longo das medies')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Legenda
legenda = [mpatches.Patch(color=v, label=k) for k, v in cor_mapa.items()]
plt.legend(handles=legenda)

plt.tight_layout()
plt.show()
```
