```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados em texto
dados = """
Tempo_ms,Temperatura_C,Status
485,6.3,NORMAL
1839,5.8,NORMAL
2839,6.7,NORMAL
3839,5.4,NORMAL
4839,8.5,NORMAL
5839,10.4,ALERTA_Pre_falha
6839,10.9,ALERTA_Pre_falha
7839,11.6,ALERTA_Pre_falha
8839,9.6,ALERTA_Pre_falha
9839,9.7,ALERTA_Pre_falha
10839,13.2,FALHA_CRITICA
11839,13.8,FALHA_CRITICA
12839,13.4,FALHA_CRITICA
13839,14.3,FALHA_CRITICA
14839,12.3,FALHA_CRITICA
15839,14.3,FALHA_CRITICA
16839,14.3,FALHA_CRITICA
17839,12.3,FALHA_CRITICA
18839,12.0,FALHA_CRITICA
19839,14.2,FALHA_CRITICA
20839,13.4,FALHA_CRITICA
21839,12.7,FALHA_CRITICA
22839,13.9,FALHA_CRITICA
23839,14.5,FALHA_CRITICA
24839,14.2,FALHA_CRITICA
25839,13.5,FALHA_CRITICA
26839,14.6,FALHA_CRITICA
27839,12.5,FALHA_CRITICA
"""

from io import StringIO
df = pd.read_csv(StringIO(dados))

# Converter Tempo_ms para string para evitar sobreposio
df["Tempo_str"] = df["Tempo_ms"].astype(str)

# Mapeamento de cores
cor_mapa = {
    'NORMAL': 'green',
    'ALERTA_Pre_falha': 'orange',
    'FALHA_CRITICA': 'red'
}
cores = df['Status'].map(cor_mapa)

# Grfico de barras
plt.figure(figsize=(16, 6))
plt.bar(df["Tempo_str"], df["Temperatura_C"], color=cores, width=0.8)

plt.xlabel("Tempo (ms)")
plt.ylabel("Temperatura (C)")
plt.title("Temperatura ao longo do tempo por status")
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Legenda
import matplotlib.patches as mpatches
legenda = [mpatches.Patch(color=v, label=k) for k, v in cor_mapa.items()]
plt.legend(handles=legenda)

plt.tight_layout()
plt.show()
```
